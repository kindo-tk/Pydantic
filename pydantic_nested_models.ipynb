# Nested models in pydantic

```python
from pydantic import BaseModel
```

```python
class Address(BaseModel):
    
    city : str  
    state : str 
    pin : str
```

```python
class Patient(BaseModel):
    
    name : str
    gender : str 
    age : int 
    address : Address
```

```python
address_dict = {"city":"bankura", "state":"WB", "pin":'722154'}
address1 = Address(**address_dict)

patient_dict = {'name':"Tufan", "gender":"male","age":23, "address": address1}

patient1 = Patient(**patient_dict)
```

```python
print(patient1)
```

```python
print(patient1.name)
```

```python
print(patient1.address.pin)
```

### Advantages of using nested models:
- Better organization of nested data
- Reusability 
- Enhanced Readability
- Validation
